```python
class TradingStrategy:
    def __init__(self, symbol : str, amount : float):
        self.symbol = symbol
        self.amount = amount
        self.position = 0
        print(id(self))
   
    
    def buy(self, price, qty):
        totalAmount = price * qty
        
        try:
            if self.amount < totalAmount:
                raise Exception("Not enough funds")
            
            self.amount -= totalAmount
            self.position += qty
            print(f"Buying {qty} of {self.symbol} at {price} Done")
            self.status()
            
        except Exception as e:
            print(f"Error: {e}")
            
        
    def sell(self, price, qty):
        totalAmount = price * qty

        try:
            if self.position < qty:
                raise Exception("Not enough shares")
            
            self.amount += totalAmount
            self.position -= qty
            
            print(f"Selling {qty} of {self.symbol} at {price} Done")
            
            self.status()
            
        except Exception as err:
            print(f"Error: {err}")
    
    def status(self):
        print(f"Symbol: {self.symbol}")
        print(f"Balance: {self.amount}")
        print(f"Position: {self.position}")
        
    
```

```python
s1 = TradingStrategy("BTC", 10000)
```

```python
s1.status()
print(id(s1))
```

```python
s1.buy(5000, 45)
```

```python
s1.buy(120, 5)
```

```python
s1.status()
```

```python
s1.sell(1200, 4)
```
