```python
class TradingStrategy:
    def __init__(self, symbol : str, amount : float):  ## Magic Methods
        self.symbol = symbol
        self.amount = amount
        self.position = 0
        print(id(self))
   
    
    def buy(self, price, qty):
        totalAmount = price * qty
        
        try:
            if self.amount < totalAmount:
                raise Exception("Not enough funds")
            
            self.amount -= totalAmount
            self.position += qty
            print(f"Buying {qty} of {self.symbol} at {price} Done")
            self.status()
            
        except Exception as e:
            print(f"Error: {e}")
            
        
    def sell(self, price, qty):
        totalAmount = price * qty

        try:
            if self.position < qty:
                raise Exception("Not enough shares")
            
            self.amount += totalAmount
            self.position -= qty
            
            print(f"Selling {qty} of {self.symbol} at {price} Done")
            
            self.status()
            
        except Exception as err:
            print(f"Error: {err}")
    
    def status(self):
        print(f"Symbol: {self.symbol}")
        print(f"Balance: {self.amount}")
        print(f"Position: {self.position}")
        
    
```

```python
s1 = TradingStrategy("BTC", 10000)
```

```python
s1.status()
print(id(s1))
```

```python
s1.buy(5000, 45)
```

```python
s1.buy(120, 5)
```

```python
s1.status()
```

```python
s1.sell(1200, 4)
```

```python
s1.status()
```

```python
class Trades:
    def __init__(self, 
                 symbol : str,
                 qty : int,
                 price : float):
        
        self.symbol = symbol
        self.qty = qty
        self.price = price
        self.price = price
        
        print(f"Executed")
    
    def __repr__(self):
        return f"Trades(symbol={self.symbol}, qty={self.qty}, price={self.price})"
    
    def __str__(self):
        return f"Trades(symbol={self.symbol}, qty={self.qty}, price={self.price})"
    
    def __add__(self, other):
        if self.symbol != other.symbol:
            return (f"Cannot excute trades for different symbols")
        
        newQty = self.qty + other.qty
        avgPrice = (self.price * self.qty + other.price * other.qty) / newQty
        print(f"The average price for {self.symbol} is {avgPrice} and the new quantity is {newQty}")
        
        return Trades(self.symbol, newQty, avgPrice)
            
        
    
```

```python
trades = Trades("BTC", 45, 5000)
```

```python
print(repr(trades))  ## representation of object in strings 

print(trades)
```

```python
print(str(trades))
```

```python
class Trader:
    def __init__(self, name : str, whichType : str):
        self.name = name
        self.whichType = whichType
    
    
    def greet(self):
        return f"Hello {self.name}, I am a {self.whichType} trader"
        
```

```python
reck98 = Trader("reck98", "Algo Trader")
```

```python
reck98.greet()
```

```python
t2 = Trader("t2", "Scalper")
```
